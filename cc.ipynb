```python
X = data.drop('default payment next month',axis=1)
```

```python
X
```

```python
y = data['default payment next month']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
X_train.head()
```

```python
#preprocessing steps
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
train_scaler = StandardScaler()
test_scaler = StandardScaler()
```

```python
scaled_train_data = train_scaler.fit_transform(X_train)
```

```python
scaled_test_data = train_scaler.transform(X_test)
```

```python
scaled_test_data
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(scaled_train_data, y_train)
```

```python
y_pred = clf.predict(scaled_test_data)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
clf2 = GaussianNB(var_smoothing=0.5)
```

```python
clf2.fit(scaled_train_data,y_train)
```

```python
y_pred = clf2.predict(scaled_test_data)
```

```python
accuracy_score(y_pred,y_test) #that is why we are now taking var_smoothing as 0.5
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
```

```python
xgb_fit = xgb.fit(scaled_train_data,y_train)
```

```python
y_pred_xgb = xgb.predict(scaled_test_data)
```

```python
accuracy_score(y_test,y_pred_xgb)
```

```python
#using hyper parameter n
```

```python
from sklearn.model_selection import GridSearchCV
param_grid_xgboost = { "n_estimators": [50,100, 130],
                        "max_depth": range(3,11,1),
                      "random_state":[0,50,100]
                     }

# Creating an object of the Grid Search class
grid= GridSearchCV(XGBClassifier(objective='binary:logistic'),param_grid_xgboost,verbose=3,cv=5,n_jobs=-1)
```

```python
grid.fit(scaled_train_data,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
xgb_new=XGBClassifier(max_depth = 7, n_estimators = 50, random_state = 0)
```

```python
y_pred4=xgb_new.fit(scaled_train_data,y_train).predict(scaled_test_data)
```

```python
accuracy_score(y_test,y_pred4)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(max_depth=2, random_state=0)
```

```python
y_pred_rfc=rfc.fit(scaled_train_data,y_train).predict(scaled_test_data)
```

```python
accuracy_score(y_test,y_pred_rfc)
```

```python
# so from the above conclusion we can say that the maximum accuracy score is comming from the random forest regressor
```
